### Preprocessing

```python
lb = LabelEncoder() 
dataset['Outlook_'] = lb.fit_transform(dataset['Outlook'])
dataset['Temperature_'] = lb.fit_transform(dataset['Temperature']) 
dataset['Humidity_'] = lb.fit_transform(dataset['Humidity']) 
dataset['Windy_'] = lb.fit_transform(dataset['Windy'])   
dataset['Play_'] = lb.fit_transform(dataset['Play'])
X = pd.DataFrame(dataset.iloc[:,5:9])
Y = dataset.iloc[:,9]
display(dataset)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=1, random_state = 32)
# display(X_train, X_test, Y_train, Y_test)
```

```python
naive_b = GaussianNB()
naive_b.fit(X_train, Y_train)
test_data = pd.concat([X_test, Y_test], axis=1)
test_data["prediction"] = naive_b.predict(X_test)
display(test_data)
print("Naive Bayes Accuracy:", naive_b.score(X_test,Y_test))
print(naive_b.predict_proba(X_test))
```
